# Bed inversion test

A twin experiment to test the bed inversion technique of van Pelt et al. (2013). The approach is to first spin a reference glacier up to steady state, then try to reproduce the basal topography based on surface "observations" (the reference model). 

# Imports

We'll use icepack's shallow stream approximation for these simulations. 

```python
import icepack, icepack.plot #for modeling ice flow
from firedrake import * #various other FEM functions
import matplotlib.pyplot as plt
import numpy as np
import tqdm #because nobody likes staring at a frozen screen with no output for minutes at a time
```

# Material parameters

Note that the default units used by icepack are a bit odd, so the quick option is to use the default values rather than input your own intuitive values.

```python
from icepack.constants import ( 
    ice_density as ρ_I, 
    water_density as ρ_W,
    gravity as g,
)

T = 255.15 #ice temperature (K)
A = icepack.rate_factor(Constant(T)) #flow parameter (weird icepack units)
```

# Meshing

```python
Lx, Ly = 10e3, 5e3 #domain length, width in metres
dx, dy = 200, 200 #mesh resolution in metres
nx, ny = int(Lx/dx), int(Ly/dy) #number of nodes

mesh = RectangleMesh(nx, ny, Lx, Ly) #2D mesh 
deg = 1 #degree of interpolation functions
Q = FunctionSpace(mesh, 'CG', deg) #function space for interpolating scalars (e.g., thickness)
V = VectorFunctionSpace(mesh, 'CG', deg) #function space for interpolating vectors (e.g., velocity)

x, y = SpatialCoordinate(mesh) #the actual spatial coords associated with the mesh
X = np.arange(0, Lx, 1) #also write a "centerline" array for plotting 1D profiles on
X = np.array([(k, Ly/2) for k in X])
```

# Design "real-life conditions"

We'll assume a bed with a linear slope, and then add a bedrock bump as a 3D cosine curve at the centre of the domain. Other choices (thickness, velocity boundary condition, etc.) will determine the initial configuration of the ice stream, which will then be modeled toward steady state before begininng the bed inversion. 

```python
H_0 = 100 #initial thickness (m)
α = -0.001 #basal slope ~ surface slope
bed = Function(Q).project(α*(x - Lx)) #note that the bed shape is laterally uniform

u_x = 10 #upstream velocity (m/a)
smb = Function(Q).project(1 - 2*x/Lx) #surface mass balance (m/a)

###############################
###############################
### create the bedrock bump ###
###############################
###############################

bump_height = H_0/10 #amplitude of bedrock bump (m)
bump_R = 1000 #radius of bump (m)

px, py = Lx/2, Ly/2 #coordinates of the center of the depression

D = conditional(
    sqrt((x - px)**2 + (y - py)**2) < bump_R,  #check if a given coord is within the radius
    1/2 * bump_height * cos(pi * sqrt((x - px)**2 + (y - py)**2) / bump_R) + 1/2 * bump_height,  #and if so, impose the cosine function here
    0  #else, zero outside the radius
)

bed.project(bed + D) #add this bump to the basal elevation
s = Function(Q).project(bed + H_0) #adjust the initial surface elevation accordingly
H = Function(Q).project(Constant(H_0)) #and the initial thickness


############
############
### plot ###
############
############

fig, axes = icepack.plot.subplots()
colors = firedrake.tripcolor(bed, axes=axes)
fig.colorbar(colors, label = 'actual bed elevation (m)')
```

# Model initialization

```python
u = Function(V).project(as_vector((u_x, 0)))
τ_b = Function(Q).project(-ρ_I*g*H*α) #estimate of basal shear stress
μ_b = Function(Q).project(τ_b/(u_x**(1/3))) #Weertman friction coefficient at the bed
μ_s = Constant(0.00) #sidewall friction coefficient - assume none for simplicity

inflow_ids = [1] #flow originates from the left boundary
sidewall_ids = [3, 4] #+/- y boundaries are free-slip sidewalls in this simulation

opts = {
    'dirichlet_ids': inflow_ids, #impose the velocity boundary condition at boundary 1
    'side_wall_ids': sidewall_ids, #since these are 3 and 4, the model automatically classifies boundary 2 as the terminus
}

model = icepack.models.IceStream(friction = icepack.models.friction.bed_friction, side_friction = icepack.models.friction.side_friction)
solver = icepack.solvers.FlowSolver(model, **opts)
```

# Spin "real" glacier up to steady state

```python
Δt = 1/2
num_years = 100 #sufficient to get to steady state
num_timesteps = int(num_years/Δt)

for step in tqdm.trange(num_timesteps):

    u = solver.diagnostic_solve( #obtain the velocity field
        velocity = u, thickness = H, surface = s, fluidity = A, friction = μ_b, side_friction = μ_s
    )

    H = solver.prognostic_solve(#account for conservative mass transport to update the thickness, given the solved velocity
        Δt, thickness = H, accumulation = smb, velocity = u, thickness_inflow = H
    )

    s = icepack.compute_surface(thickness = H, bed = bed) #update the surface elevation
    τ_b.project(-ρ_I*g*H*α) #update the basal shear
    speed = Function(Q).project(sqrt(dot(u, u))) #velocity magnitude (for updating the basal friction)
    μ_b.project(τ_b/(speed**(1/3))) #update the basal friction

fig, axes = icepack.plot.subplots()
colors = firedrake.tripcolor(s, axes=axes)
fig.colorbar(colors, label = 'elevation (m)')
axes.set_title('reference surface elevation');
```

# Try to recover the correct bed topography

In the twin experiment described in the paper, the bed was recovered by spinning PISM up from an initial ice-free configuration atop a guessed basal surface. Since icepack doesn't spin glaciers up from scratch, we will instead start from a baseline surface elevation atop a guessed basal surface. Then we'll spin up our guess model to steady state, and compare the difference in our surface vs. the "actual" surface. This will go iteratively until we find a configuration closely matching "observations." 

```python
s_ref = s.copy(deepcopy = True) #the "reference surface" is the steady-state surface just computed

bed_guess = Function(Q).project(α*(x - Lx)) #first guess for the bed (assume we know the underlying slope but do not know about the bump)
s_guess = Function(Q).project(bed_guess + H_0) #representative initial surface elevation for the upcoming model runs
H_guess = Function(Q).project(s_guess - bed_guess) #etc.
u_guess = Function(V).project(as_vector((u_x, 0))) #assume we know the velocity well enough to at least guess a boundary condition
τ_b_guess = Function(Q).project(-ρ_I*g*H_guess*α) #basal shear stress
μ_b_guess = Function(Q).project(τ_b_guess/(u_x**(1/3))) #assume weertman friction coefficient is 3

bed_evolution = [] #for later visualization
surface_evolution = []
bed_iteration_result = bed_guess.at(X, tolerance = 1e-10)
bed_evolution.append(bed_iteration_result)

K = 1/10 #obtained by trial and error
```

### The iterative process below will take much longer than the initial spinup, because we are now spinning up for each "guess" iteration along the way. 

```python
num_iterations = 50

for iteration in tqdm.trange(num_iterations):

    bed_mod = bed_guess.copy(deepcopy = True) #(re-)initialize the modeled bed from most recent guess value
    H_mod = Function(Q).project(s_guess - bed_mod) #modeled thickness...
    τ_b_mod = Function(Q).project(-ρ_I*g*H_mod*α) #etc.
    u_mod = u_guess.copy(deepcopy = True) #and this
    speed = Function(Q).project(sqrt(dot(u_mod, u_mod)))
    μ_b_mod = Function(Q).project(τ_b_mod/(speed**(1/3))) #etc

    s_mod = s_guess.copy(deepcopy = True) #but reset this

    ########################
    ### model the spinup ###
    ########################
    
    Δt = 1/2 #using the same timestep/duration choices as before
    num_years = 100
    num_timesteps = int(num_years/Δt)

    for step in range(num_timesteps):

        u_mod = solver.diagnostic_solve(
            velocity = u_mod, thickness = H_mod, surface = s_mod, fluidity = A, friction = μ_b_mod, side_friction = μ_s
        )
    
        H_mod = solver.prognostic_solve(
            Δt, thickness = H_mod, accumulation = smb, velocity = u_mod, thickness_inflow = H_mod
        )
        
        s_mod = icepack.compute_surface(thickness = H_mod, bed = bed_mod)
        τ_b_mod.project(-ρ_I*g*H_mod*α) #driving stress
        speed = Function(Q).project(sqrt(dot(u, u)))
        μ_b_mod.project(τ_b_mod/(speed**(1/3))) #weertman friction coefficient

    bed_guess.project(bed_mod - K*(s_mod - s_ref)) #update the guessed basal elevation
    bed_iteration_result = bed_guess.at(X, tolerance = 1e-10) #save current guess for later visualization
    surface_iteration_result = s_mod.at(X, tolerance = 1e-10)
    bed_evolution.append(bed_iteration_result)
    surface_evolution.append(surface_iteration_result)
```

# Plot results

```python
bed_misfit = Function(Q).project((bed_guess - bed))

fig, axes = icepack.plot.subplots()
colors = firedrake.tripcolor(bed_misfit, axes=axes)
fig.colorbar(colors, label = 'bed difference (m)')
axes.set_title('final bed misfit');
```

```python
reference_line = bed.at(X, tolerance = 1e-10)

plt.figure()
plt.title('Bed adjustment by iteration')
plt.plot(bed_guess.at(X, tolerance = 1e-10), color = 'k', linestyle = 'solid', linewidth = 1, label = 'final modeled bed')
plt.plot(bed_evolution[0], color = 'k', linestyle = 'solid', linewidth = 1, alpha = 0.1, label = 'intermediate steps')
for i in range(len(bed_evolution)):
    plt.plot(bed_evolution[i], color = 'k', linestyle = 'solid', linewidth = 1, alpha = 0.1)
plt.plot(reference_line, color = 'k', linestyle = 'dotted', linewidth = 3, label = 'actual bed')
plt.xlabel('longitudinal coordinate (m)')
plt.ylabel ('elevation (m)')
# plt.grid()
plt.legend()
plt.show()
```
